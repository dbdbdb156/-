## 2.3.4 나이브 베이즈 분류기
= (비슷) 선형 모델

 #### ※ 특징
  
  - 각 특성을 개별로 취급 => 클래스별 통계를 단순하게 취합
 
 
 #### ※ 종류
  
  
  - GaussianNB
      : 연속적인 어떤 데이터에도 적용가능
  
  - BernoulliNB
      : 이진데이터
  
  - MultinomialNB
      : 카운트 데이터에 적용(Counter data : 특성이 어떤 것을 헤아린 정수 카운트)
      

  

```python
## import 를 다 해놓는 겁니다 !
%matplotlib inline
from preamble import *
```

```python
X = np.array([[0, 1, 0, 1],
              [1, 0, 1, 1],
              [0, 0, 0, 1],
              [1, 0, 1, 0]])
y = np.array([0, 1, 0, 1])
```

```python
counts = {}
for label in np.unique(y):
    # 각 클래스에 대해 반복
    # 특성마다 1 이 나타난 횟수를 센다.
    counts[label] = X[y == label].sum(axis=0)
print("특성 카운트:\n{}".format(counts))
```

#### ※ 다른 두 나이브 베이즈 모델의 차이
: 계산하는 통계 데이터의 종류가 다름
    
- MultinomialNB : 클래스별 특성의 평균 계산

- GaussianNB : 클래스별 특성의 표준편차와 평균

※ 나이브 베이즈 모델에서는 coef  ! = w (가중치)  가 아니다 다르다 !

### 장,단점와 매겨변수

- 가상의 데이터 푄트를 alpha 갯수만큼 추가
    => 통계 데이터 완만

- GaussianNB < - > 다른 두 나이브 베이즈 모델 : 텍스트 같은 희소한 데이터 카운트
 
 1) 매우 고차원 데이터셋에 사용
 
 장점 = 선형모델과 장단점이 비슷
 
     - 훈련과 예측 속도가 빠름
